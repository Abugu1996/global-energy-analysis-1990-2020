 Standardize column names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace('/', '_').str.replace(r"[()%-]", "", regex=True)
```

```python
df.head()
```

 Remove duplicates

```python
df = df.drop_duplicates()
```

```python
df.shape
```

 Convert year to int

```python
if "year" in df.columns:
    df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")
```

```python
df.dtypes
```

 Convert numeric columns

```python
for col in df.select_dtypes(include="object").columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except (ValueError, TypeError):
        print(f"Skipped column: {col}")

```

Describe and clean the numeric columns

```python
df_numeric = df.select_dtypes(include=['number'])
numeric_desc = df_numeric.describe().dropna(how='all', axis=0)


```

Describe and clean the object columns

```python
df_object = df.select_dtypes(include=['object'])
object_desc = df_object.describe().dropna(how='all', axis=0)
```

Separate numeric and object columns

```python
df_numeric = df.select_dtypes(include=['number'])
df_object = df.select_dtypes(include=['object'])
```

```python
numeric_desc = df_numeric.describe()
object_desc = df_object.describe()
```

```python
print("--- Numeric Summary ---")
print(numeric_desc)
print("\n--- Object Summary ---")
print(object_desc)
```

 Histograms

```python
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols].hist(figsize=(15,10), bins=20)
plt.suptitle("Distribution of Numeric Variables")
plt.savefig(OUT_DIR / "eda_distributions.png", dpi=150)
plt.show()
```

Correlation heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.savefig(OUT_DIR / "eda_heatmap.png", dpi=150)
plt.show()
```

```python
col_total = "total_energy_consumption_mtoe"
col_renew = "share_of_renewables_in_electricity_production"
col_elec = "electricity_production_twh"
col_country = "country"
col_year = "year"
```

Total Energy Consumption by Year

```python
kpi_yearly = df.groupby(col_year)[col_total].sum().reset_index()
kpi_yearly.to_csv(OUT_DIR / "kpi_total_energy_by_year.csv", index=False)
plt.figure(figsize=(10,5))
plt.plot(kpi_yearly[col_year], kpi_yearly[col_total], marker="o")
plt.title("Total Energy Consumption by Year")
plt.xlabel("Year")
plt.ylabel("Energy (Mtoe)")
plt.grid(True)
plt.savefig(OUT_DIR / "kpi_total_energy_by_year.png", dpi=150)
plt.show()
```

Total Energy Consumption by Year

```python
kpi_total_by_year = (
    df.groupby("year")["total_energy_consumption_mtoe"]
      .sum()
      .reset_index()
)
```

```python
kpi_total_by_year.to_csv(OUT_DIR / "kpi_total_energy_by_year.csv", index=False)

plt.figure(figsize=(10,5))
plt.plot(kpi_total_by_year["year"], kpi_total_by_year["total_energy_consumption_mtoe"], marker="o")
plt.title("Total Energy Consumption by Year")
plt.xlabel("Year")
plt.ylabel("Energy (Mtoe)")
plt.grid(True)
plt.savefig(OUT_DIR / "kpi_total_energy_by_year.png", dpi=150)
plt.show()
```

Average & Weighted Renewable Energy Share

```python
df["weighted_renewable_share"] = (
    df["share_of_renewables_in_electricity_production_"] * df["electricity_production_twh"]
)
```

```python

avg_renew = df["share_of_renewables_in_electricity_production_"].mean()
weighted_renew = df["weighted_renewable_share"].sum() / df["electricity_production_twh"].sum()

print(f"Average Renewable Share: {avg_renew:.2f}%")
print(f"Weighted Renewable Share: {weighted_renew:.2f}%")

pd.DataFrame({
    "Average_Renewable_Share": [avg_renew],
    "Weighted_Renewable_Share": [weighted_renew],
    "Weighted_Non_Renewable": [100 - weighted_renew]
}).to_csv(OUT_DIR / "kpi_renew_vs_nonrenew.csv", index=False)
```

Pie chart

```python
plt.figure(figsize=(6,6))
plt.pie(
    [weighted_renew, 100-weighted_renew],
    labels=["Renewable", "Non-Renewable"],
    autopct="%.1f%%",
    colors=["green","gray"]
)
plt.title("Global Renewable vs Non-Renewable Share")
plt.savefig(OUT_DIR / "renew_vs_nonrenew_pie.png", dpi=150)
plt.show()
```

Top 10 Countries by Energy Consumption

```python
top_countries = (
    df.groupby("country")["total_energy_consumption_mtoe"]
      .sum()
      .reset_index()
      .sort_values(by="total_energy_consumption_mtoe", ascending=False)
      .head(10)
)
top_countries.to_csv(OUT_DIR / "kpi_top_countries.csv", index=False)

plt.figure(figsize=(10,6))
sns.barplot(x="total_energy_consumption_mtoe", y="country", data=top_countries, hue="country", palette="viridis", legend=False  )
plt.title("Top 10 Countries by Total Energy Consumption")
plt.savefig(OUT_DIR / "kpi_top_countries.png", dpi=150)
plt.show();
```

Growth Rate in Energy Consumption

```python
kpi_total_by_year["yoy_pct"] = kpi_total_by_year["total_energy_consumption_mtoe"].pct_change() * 100

first_val = kpi_total_by_year["total_energy_consumption_mtoe"].iloc[0]
last_val = kpi_total_by_year["total_energy_consumption_mtoe"].iloc[-1]
n_years = kpi_total_by_year["year"].iloc[-1] - kpi_total_by_year["year"].iloc[0]
cagr = (last_val / first_val) ** (1/n_years) - 1

print(f"CAGR ({kpi_total_by_year['year'].iloc[0]}–{kpi_total_by_year['year'].iloc[-1]}): {cagr*100:.2f}%")

kpi_total_by_year.to_csv(OUT_DIR / "kpi_growth.csv", index=False)

plt.figure(figsize=(10,5))
plt.plot(kpi_total_by_year["year"], kpi_total_by_year["yoy_pct"], marker="o", color="red")
plt.title("YoY Growth Rate in Energy Consumption")
plt.xlabel("Year")
plt.ylabel("YoY % Change")
plt.grid(True)
plt.savefig(OUT_DIR / "kpi_growth.png", dpi=150)
plt.show()




```

CO2 Intensity Trend

```python
kpi_co2_intensity = (
    df.groupby("year")["average_co2_emission_factor_tco2_toe"]
      .mean()
      .reset_index()
)
kpi_co2_intensity.to_csv(OUT_DIR / "kpi_co2_intensity.csv", index=False)

plt.figure(figsize=(10,5))
plt.plot(kpi_co2_intensity["year"], kpi_co2_intensity["average_co2_emission_factor_tco2_toe"], marker="o", color="orange")
plt.title("Average CO₂ Emission Factor Trend")
plt.xlabel("Year")
plt.ylabel("tCO₂ per TOE")
plt.grid(True)
plt.savefig(OUT_DIR / "kpi_co2_intensity.png", dpi=150)
plt.show()
```

Share of Electricity in Final Energy Consumption

```python
kpi_elec_share = (
    df.groupby("year")["share_of_electricity_in_total_final_energy_consumption_"]
      .mean()
      .reset_index()
)
kpi_elec_share.to_csv(OUT_DIR / "kpi_electricity_share.csv", index=False)

plt.figure(figsize=(10,5))
plt.plot(kpi_elec_share["year"], kpi_elec_share["share_of_electricity_in_total_final_energy_consumption_"], marker="o", color="blue")
plt.title("Share of Electricity in Final Energy Consumption")
plt.xlabel("Year")
plt.ylabel("Share (%)")
plt.grid(True)
plt.savefig(OUT_DIR / "kpi_electricity_share.png", dpi=150)
plt.show()
```
